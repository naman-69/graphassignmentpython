```python
#Q1
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

y = x**3

plt.plot(x, y, label='y = x^3')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
#Q2
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
#Q3
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1.5*np.pi, 1.5*np.pi, 400)
x = x[~np.isclose(np.abs(x) % (np.pi), np.pi / 2)]

y_tan = np.tan(x)
y_cot = 1 / np.tan(x)

plt.plot(x, y_tan, label='tan(x)')
plt.plot(x, y_cot, label='cot(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
#Q5
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)

y_square = x**2
y_cube = x**3

plt.plot(x, y_square, label='y = x^2')
plt.plot(x, y_cube, label='y = x^3')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(bbox_to_anchor =(0.75, 1.15), ncol = 2, loc ="lower right")

plt.show()
```

```python
#6 , 7 and 8th question are done in one

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)

y_square = x**2
y_cube = x**3

plt.plot(x, y_square, label='y = x^2')
plt.plot(x, y_cube, label='y = x^3')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2 and y = x^3')

plt.legend(bbox_to_anchor =(0.75, 1.15), ncol = 2, loc ="lower right")

plt.show()
```

```python
#Q9
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)

y_square = x**2
y_cube = x**3

plt.scatter(x, y_square, color='blue', label='y = x^2', marker='o')

plt.scatter(x, y_cube, color='orange', label='y = x^3', marker='x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of y = x^2 and y = x^3')

plt.legend()

plt.show()
```

```python
#Q10
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)

y_square = x**2
y_cube = x**3

plt.scatter(x, y_square, color='blue', label='y = x^2', marker='o', linestyle='-')

plt.scatter(x, y_cube, color='orange', label='y = x^3', marker='x', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of y = x^2 and y = x^3')

plt.legend()

plt.show()
```
